```
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```
#스캐터 relation
p = figure(plot_width=600, plot_height=400)
```

```
show(p)
```

```
x = [0,1,2,3,4,5]
y = [0,1,2,3,4,5]
p.circle(x,y, size = 10)
show(p)
```

```
p.circle(x,y, size = 20, line_color='blue', fill_color='#FF1493')
show(p) # 칼라정하기 bokeh python circle 검색, https://www.w3schools.com/colors/colors_groups.asp
```

```
p.circle(x,y, size = 20, line_color='blue', fill_color='#FF1493', fill_alpha=0.2) #알파값주기
show(p)
```
