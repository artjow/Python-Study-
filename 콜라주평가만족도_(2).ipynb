```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc
from matplotlib import style

style.use('ggplot')

industry = ['Hue', 'Texture', 'Harmony', 'Creativity', 'Impression', 'Abstract reality', 'Reminiscent', 'Synthesis', 'Photoreality', 'Statement']
fluctuations = [30, 65, 74, 36,25, 55, 60,65,20,40]

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)

ypos = np.arange(10)
rects = plt.barh(ypos, fluctuations, align='center', alpha=0.7, color='red', height=0.5)
plt.yticks(ypos, industry)

plt.ylabel('(Collage evaluation items)')
plt.title('< Collage satisfaction > ')

for i, rect in enumerate(rects):
    ax.text(0.95 * rect.get_width(), rect.get_y() + rect.get_height() / 2.0, str(fluctuations[i]) + '%', ha='right', va='center')

plt.xlabel('Percent(%)')
plt.show()
```
