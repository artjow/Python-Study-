```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
tips=sns.load_dataset('tips')
tips
```

```
sns.set(style='darkgrid')#바탕색 지정
sns.relplot(x='total_bill', y='tip', data=tips)
```

```
#hue 색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```
#사이즈
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```
#색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='GnBua')
```

```
#색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='GnBu')
```

```
#사이즈
#색상
sns.relplot(x='total_bill', y='tip', data=tips, size='size')
```

```
df=pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500). cumsum()))#cumsum누적합
df
```
