```
'''
iris라는 데이터를 가지고 오도록 하겠습니다. 
이 데이터는 꽃 아이리스의 종별 꽃잎과 꽃받침에 대한 자료를 가지고 있습니다. 
seaborn의 그래픽 기능을 확인하는 용도로 보도록 하겠습니다.
'''

import pandas as pd
import numpy as np
import seaborn as sns; sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")
iris.head(10)
```

```
'''
pairplot으로 
각 column별 데이터에 대한 상관관계나 분류적 특성을 
바로 명령 한 줄로 확인
'''
sns.pairplot(iris)
plt.show()
```

```
#hue 옵션을 이용해서 종별 특성을 색상으로도 알 수 있습니다.
sns.pairplot(iris, hue="species")
plt.show()
```

```
#팔렛트를 이용해서 색상 셋을 바꿀 수 있다.
sns.pairplot(iris, hue="species", palette="husl")
plt.show()
```

```
#몇개의 column만 따로 볼 수 있습니다.
sns.pairplot(iris, size=3, vars=["sepal_width", "sepal_length"])
plt.show()
```

```
'''
jointplot은
scatter 플롯을 보이면서도 또 각 축별로 히스토그램을 보여주고 있습니다. 
특히 상관관계중 피어슨비율을 함께 보여주고 있습니다.
'''
tips = sns.load_dataset("tips")
sns.set_style("whitegrid")
sns.jointplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```
#kde로 표현할 수 있습니다.
iris = sns.load_dataset("iris")
sns.jointplot("sepal_width", "petal_length", data=iris, kind="kde", space=0, color="g")
plt.show()
```
